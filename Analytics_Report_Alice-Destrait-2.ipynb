```python
def process_dataframe(df):
    # Check if all values in each column are null or empty
    empty_columns = [col for col in df.columns if df[col].isnull().all()]

    # Delete empty columns from the DataFrame
    df.drop(columns=empty_columns, inplace=True)

    # Find empty rows and delete them
    df.drop(df[df.isnull().all(axis=1)].index, inplace=True)

    # Complete duplicated rows
    df.drop_duplicates(inplace=True)

    # Reset index after dropping rows
    df.reset_index(drop=True, inplace=True)
    print(df.shape)

process_dataframe(meta)
process_dataframe(produ)
```

## Problem 1

```python
# Calculate the number of bins needed based
num_bins_needed = int(np.ceil(meta['CumGas_MCF'].max() / 100000))
num_bins = 40 # Better overview in the graph
# Adjust the number of bins if needed
if num_bins_needed < 40:
    num_bins = num_bins_needed

# Create a new column with bin labels
meta['CumGas_MCF_Bin'] = pd.cut(meta['CumGas_MCF'], bins=np.arange(0, (num_bins + 1) * 100000, 100000))

# Plot the crosstab with bins
pd.crosstab(meta['CumGas_MCF_Bin'], meta['County']).plot(kind='bar', figsize=(10, 6))
plt.title('Cumulative gas per county')
plt.xlabel('Gas')
plt.ylabel('County')
plt.savefig('data/image.png')

plt.show()
```

```python
# Calculate the number of bins needed based on the bin width
num_bins_needed = int(np.ceil(meta['CumWater_BBL'].max() / 100000))

num_bins = 40 # Better overview in the graph
# Adjust the number of bins if needed
if num_bins_needed < 40:
    num_bins = num_bins_needed

# Create a new column with bin labels
meta['CumWater_BBL_Bin'] = pd.cut(meta['CumWater_BBL'], bins=np.arange(0, (num_bins + 1) * 100000, 100000))

# Plot the crosstab with bins
pd.crosstab(meta['CumWater_BBL_Bin'], meta['County']).plot(kind='bar', figsize=(10, 6))
plt.title('Cumulative water per county')
plt.xlabel('Water')
plt.ylabel('County')
plt.savefig('data/image-2.png')

plt.show()
```

```python
# Calculate the number of bins needed based on the bin width
num_bins_needed = int(np.ceil(meta['CumOil_BBL'].max() / 50000))
num_bins = 30 # Smaller amounts than water or gas
# Adjust the number of bins if needed
if num_bins_needed < 30:
    num_bins = num_bins_needed

# Create a new column with bin labels
meta['CumOil_BBL_Bin'] = pd.cut(meta['CumOil_BBL'], bins=np.arange(0, (num_bins + 1) * 50000, 50000))

# Plot the crosstab with bins
pd.crosstab(meta['CumOil_BBL_Bin'], meta['County']).plot(kind='bar', figsize=(10, 6))
plt.title('Cumulative oil per county')
plt.xlabel('Oil')
plt.ylabel('County')
plt.savefig('data/image-3.png')

plt.show()
```

```python
# Overview of which well type is better
welltype = meta.groupby('ENVProdWellType')['EURWH_MBOE'].mean().round(2).sort_values(ascending=False)

# Create a DataFrame to display the ranking
ranking_df = pd.DataFrame({'Well Types': welltype.index,
                            'Estimated Ultimate Recovery (BOE)': welltype.values})

print(ranking_df)
```

```python
# Overview of which producing method is better
producing_meth = meta.groupby('ENVProducingMethod')['EURWH_MBOE'].mean().round(2).sort_values(ascending=False)

# Create a DataFrame to display the ranking
ranking_df = pd.DataFrame({'Producing Method': producing_meth.index,
                            'Estimated Ultimate Recovery (BOE)': producing_meth.values})

print(ranking_df)
```

```python
# Overview of which operator has the best results
operator_production = produ.groupby('ProducingOperator')['Prod_BOE'].mean().round(2).sort_values(ascending=False)

# Create a DataFrame to display the ranking
ranking_df = pd.DataFrame({'Producing Operator': operator_production.index,
                            'Average Production (BOE)': operator_production.values})

# Display the top performers performers table
print(ranking_df[ranking_df['Average Production (BOE)'] > 20000])
```

```python
# Total cost compared to the location of the well
min_lat = meta['Latitude'].min()
max_lat = meta['Latitude'].max()
min_lon = meta['Longitude'].min()
max_lon = meta['Longitude'].max()

# Create a map to visualize well locations and costs, focusing only on the concentrated area
m = folium.Map(location=[(min_lat + max_lat) / 2, (min_lon + max_lon) / 2], zoom_start=10)

# Add markers for well locations with TotalWellCost_USDMM comparison
for index, row in meta.iterrows():
    darkness = min(1, row['TotalWellCost_USDMM'] / meta['TotalWellCost_USDMM'].max())
    
    red_value = int(darkness * 255)
    green_value = 0
    blue_value = 0
    
    # Construct RGB color
    color_hex = '#{0:02x}{1:02x}{2:02x}'.format(red_value, green_value, blue_value)
    
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=5 + row['TotalWellCost_USDMM'] / 10,  # Adjust radius based on TotalWellCost_USDMM
        tooltip=row['WellName'],
        color=color_hex,  # Marker color based on darkness
        fill=True,
        fill_color=color_hex,  # Fill color based on darkness
        fill_opacity=0.6
    ).add_to(m)

m
```

```python
# Overview of how variables correlate to the total cost
columns_of_interest = ['TotalWellCost_USDMM', 'EURWH_MBOE', 'PeakOil_BBL', 'PeakGas_MCF',
                       'CumOil_BBL', 'CumGas_MCF', 'TotalProducingMonths', 'MD_FT']

# Creating a subset DataFrame with selected columns
subset_data = meta[columns_of_interest]

corr = subset_data.corr()

# Plotting heatmap for correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], 
            annot=True, 
            cmap='viridis', 
            fmt=".2f", 
            vmax=1.0,
            vmin=-1.0,
            mask = np.triu(np.ones_like(corr, dtype=bool)),
            annot_kws={"size": 10})
plt.title('Correlation Matrix of Production Metrics and Total Well Cost', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```
